```python
#remove the old profiles, all of which either have "normalized" or "augmented" in the name
import os
import glob
aug = glob.glob('*/*/*/*_augmented.csv.gz')
for eachaug in aug:
    os.remove(eachaug)
norm = glob.glob('*/*/*/*_normalized*')
for eachnorm in norm:
    os.remove(eachnorm)
```

# update metadata

```python
# load in new metadata file, add missing mmoles_per_liter, not anything else
import pandas
MOA = pandas.read_csv('metadata/JUMP-MOA_compound_platemap_with_metadata.csv')
MOA['mmoles_per_liter']=1.5
MOA['pert_type']='trt'
MOA['control_type']=''
MOA.fillna('', inplace=True)
MOA.loc[(MOA['moa'] == ''),['mmoles_per_liter']]=''
MOA.loc[(MOA['moa'] == ''),['pert_type']]='control'
MOA.loc[(MOA['moa'] == ''),['control_type']]='negcon'
```

```python
# find plates that are using the old name for the JUMP-MOA plate
MOA_batches = []
non_MOA_batches = []
batchlist = os.listdir('profiles')
for eachbatch in batchlist:
    file_to_find = os.path.join('metadata','platemaps',eachbatch,'platemap','SANITIZED.txt')
    if os.path.exists(file_to_find):
        os.remove(file_to_find)
        MOA_batches.append(eachbatch)
        MOA.to_csv(os.path.join('metadata','platemaps',eachbatch,'platemap','JUMP-MOA_compound_platemap_with_metadata.txt'),index=False, sep='\t')
        platemap_name = os.path.join('metadata','platemaps',eachbatch,'barcode_platemap.csv')
        barcode_platemap = pandas.read_csv(platemap_name)
        barcode_platemap['Plate_Map_Name'] ='JUMP-MOA_compound_platemap_with_metadata'
        barcode_platemap.to_csv(platemap_name,index=False)
        
    else:
        if '.' not in eachbatch:
            non_MOA_batches.append(eachbatch)

print(non_MOA_batches)

```

```python
batches_and_plates={}
for eachbatch in MOA_batches:
    batchfolder = os.path.join('profiles',eachbatch)
    platelist = os.listdir(batchfolder)
    for eachplate in platelist:
        per_plate_csv = os.path.join(batchfolder,eachplate,eachplate+'.csv.gz')
        if os.path.exists(per_plate_csv):
            if eachbatch not in batches_and_plates.keys():
                batches_and_plates[eachbatch]=[eachplate]
            else:
                batches_and_plates[eachbatch]+=[eachplate]
```

```python
batches_for_config = list(batches_and_plates.keys())
batches_for_config.sort()
for eachbatch in batches_for_config:
    print("---")
    print(f"batch: {eachbatch}")
    print('plates:')
    for eachplate in batches_and_plates[eachbatch]:
        print(f"  - name: {eachplate}")
        print("    process: true")
    print("process: true")
```
