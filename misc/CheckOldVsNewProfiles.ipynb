```python
import os
import pandas
pandas.__version__
```

```python
old_profile_path = '/Users/bcimini/Documents/GitHub/jump-cellpainting/pilot-data/profiles/'
new_profile_path = '/Users/bcimini/work/projects/workspace/software/pilot-data-public/profiles/'
```

```python
batchlist = os.listdir(new_profile_path)
batchlist = [x for x in batchlist if '.' not in x]
batchlist.sort()
compare_df = pandas.DataFrame()
compare_list = []
master_missing_column_list = []
batchcount = len(batchlist)
platecount = 0
columncount = 0
for eachbatch in batchlist:
    print("")
    print(f"Starting assessment of {eachbatch}")
    platelist = os.listdir(os.path.join(new_profile_path,eachbatch))
    platelist.sort()
    for eachplate in platelist:
        full_plate_name = os.path.join(new_profile_path,eachbatch,eachplate,eachplate+'_normalized_feature_select_negcon.csv.gz')
        if os.path.exists(full_plate_name):
            full_old_plate_name = False
            full_old_plate_name_1 = os.path.join(old_profile_path,eachbatch,eachplate,eachplate+'_normalized_feature_select_negcon.csv.gz')
            full_old_plate_name_2 = os.path.join(old_profile_path,eachbatch,eachplate,eachplate+'_normalized_feature_select_dmso.csv.gz')
            if os.path.exists(full_old_plate_name_1):
                full_old_plate_name = full_old_plate_name_1
            elif os.path.exists(full_old_plate_name_2):
                full_old_plate_name = full_old_plate_name_2
            if not full_old_plate_name:
                print(f"Could not find corresponding old plate for {eachplate}")
            else:
                platecount +=1
                df_new = pandas.read_csv(full_plate_name)
                df_old = pandas.read_csv(full_old_plate_name)
                missing_column_list_1 = [x for x in df_new.columns if 'Metadata' not in x if x not in df_old.columns]
                missing_column_list_2 = [x for x in df_old.columns if 'Metadata' not in x if x not in df_new.columns]
                missing_column_list = missing_column_list_1 + missing_column_list_2
                if len(missing_column_list) == 0:
                    print(f"No missing columns in the plate {eachplate}")
                else:
                    if eachbatch != '2021_04_20_Target2': #This batch we know was selected differently
                        for eachcol in missing_column_list_1:
                            master_missing_column_list.append(eachcol+'_is_missing_in_new_profiles_from_batch_'+eachbatch+'_plate_'+eachplate)
                        for eachcol in missing_column_list_2:
                            master_missing_column_list.append(eachcol+'_is_missing_in_old_profiles_from_batch_'+eachbatch+'_plate_'+eachplate)
                        print(f"The list of columns present in the new dataframe but not the old dataframe is {missing_column_list_1} ")
                        print(f"The list of columns present in the old dataframe but not the new dataframe is {missing_column_list_2} ")
                    else:
                        print(f"Missing columns expected for {eachplate} due to change in selection level (plate vs batch)")
                different_columns = []
                for column in df_new.columns:
                    if 'Metadata' not in column:
                        if column not in missing_column_list:
                            columncount += 1
                            try: 
                                pandas.testing.assert_series_equal(df_new[column],df_old[column])
                            except AssertionError:
                                different_columns.append(column)
                                compare_list.append(eachbatch+'_'+eachplate+'_'+column)
                                compare_df[eachbatch+'_'+eachplate+'_'+column+'_old']= df_old[column]
                                compare_df[eachbatch+'_'+eachplate+'_'+column+'_new']= df_new[column]
                if len(different_columns) == 0:
                    print(f"Old and new data frames for plate {eachplate} are identical to rtol=1e-05, atol=1e-08")
                else:
                    print(f"The following list of columns for plate {eachplate} are not identical; this should be followed up further. {different_columns}")
                
                      
```

```python
compare_df.fillna('')
diffcount = 0
ratio_list = []
for eachcomparison in compare_list:
    sub_df = compare_df.query(eachcomparison+'_old != '+eachcomparison+'_new')
    print(f'Ratio of new to old values in batch/plate/column {eachcomparison} that are NOT equal, with new and old values')
    for _, row in sub_df.iterrows(): 
        diffcount +=1
        try:
            ratio = float(row[eachcomparison+'_new'])/float(row[eachcomparison+'_old'])
            if pandas.notna(ratio):
                ratio_list.append(ratio)
        except:
            ratio='nan'
        print(ratio,float(row[eachcomparison+'_new']),float(row[eachcomparison+'_old']))
    print('')
```

```python
print(f"The entire missing column list is {master_missing_column_list}")
print("")
print(f"In total, across {columncount} common non-metadata columns assessed from {platecount} plates in {batchcount} batches, {diffcount} values from {len(compare_list)} columns were different")
print(f"This represents an approximate difference rate of {(diffcount*100)/(384*columncount)} %")
```

```python
import numpy
print(ratio_list)
ratio_list = numpy.array(ratio_list)
print(min(ratio_list))
print(numpy.mean(ratio_list))
print(numpy.median(ratio_list))
print(max(ratio_list))
```
